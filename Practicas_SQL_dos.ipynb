```python
query = '''  

SELECT
  TC.id_cliente,
  TC.nombre AS nombre_cliente,
  TC.ciudad,
  TC.grado,
  TS.id_vendedor,
  TS.ciudad
FROM
  tabla_customer TC
LEFT JOIN
  tabla_salesman TS ON TC.id_vendedor = TS.id_vendedor
ORDER BY
  TC.id_cliente ASC

 '''
    
cursor.execute(query)
conexion.commit()
cols = [description[0] for description in cursor.description]
pd.DataFrame(cursor.fetchall(), columns=cols)
```
